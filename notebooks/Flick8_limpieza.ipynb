## 2. Procesamiento de imágenes para normalización

**Importamos librerías relevantes**

```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from skimage.transform import rescale
from skimage import img_as_ubyte
import pickle
```

**Utilidades necesarias para normalizar las imágenes**

```
def margin_img(I):

  '''
  Funcion que devuelve una version normalizada de una imagen
  '''
  H, W = I.shape[0], I.shape[1]
  r = 256 / np.max([H, W])
  I = rescale(I, (r, r, 1))
  O = np.zeros((256, 256, 3))
  
  H, W = I.shape[0], I.shape[1]
  margin = np.round(np.abs(H-W) / 2).astype('int')
  if H > W:
    O[:, margin:margin+W, :] = I
  else:
    O[margin:margin+H, :, :] = I
        
  return img_as_ubyte(O)
```
